```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Airline Quality Ratings.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Drop the null values
df.drop(columns="Arrival Delay",axis=0,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# Example: Converting the "Satisfaction" column to numeric
# Replace 'Satisfied' with 1 and 'Dissatisfied' with 0 (modify as needed)
df['Satisfaction'] = df['Satisfaction'].map({'Satisfied': 1, 'Dissatisfied': 0})

# Check the result
print(df['Satisfaction'].head())
```

```python
df.info()
```

```python
# Replace null values in the "Satisfaction" column with a default value (e.g., 0)
df['Satisfaction'] = df['Satisfaction'].fillna(0)
```

```python
df.drop(i, axis = 0, inplace = True)
```

```python
#Seperate the coumns into x & y
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x.shape
```

```python
y.shape
```

```python
#Encoding
cat_col = x.select_dtypes(object).columns
```

```python
cat_col
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
x[cat_col] = oe.fit_transform(x[cat_col])
```

```python
x
```

```python
# Split the data into training & testing 
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.3,random_state = 1)
```

```python
#1)Predicting the data using logistic Regression
from sklearn.linear_model import LogisticRegression
logreg= LogisticRegression()
logreg.fit(xtrain,ytrain)
ypred =logreg.predict(xtest)
```

```python
#Evaluate the model
from sklearn.metrics import accuracy_score,classification_report
ac = accuracy_score(ytest,ypred)
cr = classification_report(ytest,ypred)
print("Accuracy score : ",ac)
print(cr)
```

```python
#HPT
logreg = LogisticRegression(solver = "liblinear")
logreg.fit(xtrain,ytrain)
ypred = logreg.predict(xtest)
```

```python
ac = accuracy_score(ytest,ypred)
cr = classification_report(ytest,ypred)
print("Accuracy score : ",ac)
print(cr)
```

```python
logreg = LogisticRegression(solver = 'newton-cholesky')
logreg.fit(xtrain,ytrain)
ypred = logreg.predict(xtest)
```

```python
ac = accuracy_score(ytest,ypred)
cr = classification_report(ytest,ypred)
print("Accuracy score : ",ac)
print(cr)
```

```python
#2)Predicting the data using KNN Classifier
# by default n_neighbors = 5
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5) # by default n_neighbors = 5
knn.fit(xtrain,ytrain)
ypred = knn.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
ac = accuracy_score(ytest,ypred)
print(ac)
```

```python
#3)Predicting the data using Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
dt = DecisionTreeClassifier()
```

```python
def mymodel(model):
    model.fit(xtrain,ytrain)
    ypred = model.predict(xtest)
    print(accuracy_score(ytest,ypred))
    print(classification_report(ytest,ypred))
    
    return model
```

```python
mymodel(dt)
```

```python
from sklearn import tree
fig,ax = plt.subplots(figsize = (20,10))
chart = tree.plot_tree(dt,max_depth = 5,feature_names = x.columns,filled = True,fontsize=10)
plt.savefig("DecisionTree.jpg")
```

```python
train = dt.score(xtrain,ytrain)
test = dt.score(xtest,ytest)
print(f"train score :{train} \n test score : {test}")
```

```python
#Hyper Tunning the Decision Tree
#1]max_depth
for i in range(30,45):
    dt1 = DecisionTreeClassifier(max_depth = i)
    dt1.fit(xtrain,ytrain)
    ypred = dt1.predict(xtest)
    ac = accuracy_score(ytest,ypred)
    print(f"max_depth = {i} accuracy : {ac}")
```

```python
dt2 = DecisionTreeClassifier(max_depth =44)
mymodel(dt2)
```

```python
train = dt2.score(xtrain,ytrain)
test = dt2.score(xtest,ytest)
print(f"train score :{train} \n test score : {test}")
```

```python
#2]min_samples_split
for i in range(80,91):
    dt3 = DecisionTreeClassifier(min_samples_split = i)
    dt3.fit(xtrain,ytrain)
    ypred = dt3.predict(xtest)
    ac = accuracy_score(ytest,ypred)
    print(f"max_depth = {i} accuracy : {ac}")
```
